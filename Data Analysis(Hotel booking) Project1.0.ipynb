# Data Analysis and visualization

```python
cancelled_prec=data['is_canceled'].value_counts(normalize=True)
cancelled_prec


plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not cancelled','cancelled'],data['is_canceled'].value_counts())
```

# 62%- not cancel, 37%- cancel


The accompanying bar graph shows the percentage of reservation that are canceled and those that are not .It is obvious that there are still a significant number of reservation that have not been canceled.There are still 37% of clients who canceled their reservation which has as significant impact on the hotels earning

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=data,palette='Blues')
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

# 0 for not canceled, 1 for canceled

In comparison to resort hotels ,city hotels have more bookings.It`s possible that resort hotels are more expensive than those in cities.

```python
resort_hotel=data[data['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=data[data['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize=30)
sns.lineplot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='city Hotel')
plt.legend(fontsize=20)
plt.show()
```

The line graph above shows that ,on certain days, the average daily rate for a city hotel is less than of a resort hotel,and on other days, it is even less.It goes without saying that weekends and holidays may see a rise in resort hotel rates.

```python
data['month']=data['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=data,palette='bright')
plt.title('Resevation status per month',size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
sns.barplot('month','adr',data=data[data['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.legend(fontsize=20)
plt.show()
```

```python
cancelled_data=data[data['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

In this pie chart you can see in Portugal cpuntry has high rate of cancellation is 70% as compare to other country has less rate of cancellation. So if we want to give suggestion to hotel, we can give to Portugal country.They should give good facilities to their coustomer,price should be affordable,promotional discount should be given,campaign should run,advertisement,marketing they should do to run their business.Then only they rediuce the cancellation rate.
